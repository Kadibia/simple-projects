```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.target_names)
df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)
```

```python
len(X_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier() #trained with default tuned nunmber of tree `n_estimator=100`
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=10) #trained with default tuned nunmber of tree `n_estimator=100`
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
