```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
df
```

```python
df.drop(['sepal length (cm)', 'sepal width (cm)'], axis=1, inplace=True)
```

```python
df
```

```python
plt.scatter(df['petal length (cm)'], df['petal width (cm)'])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[['petal length (cm)', 'petal width (cm)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head()
```

```python
df.max()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='green')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], color='red')
plt.scatter(df3['petal length (cm)'], df3['petal width (cm)'], color='black')

plt.xlabel('Petal Length {cm}')
plt.ylabel('Petal width {cm}')
plt.legend()
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['petal width (cm)']])
df['petal width (cm)'] = scaler.transform(df[['petal width (cm)']])
df

scaler = MinMaxScaler()
scaler.fit(df[['petal length (cm)']])
df['petal length (cm)'] = scaler.transform(df[['petal length (cm)']])
df
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['petal length (cm)', 'petal length (cm)']])
y_predicted
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='green', label='Petal width 1')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], color='red', label='Petal width 2')
plt.scatter(df3['petal length (cm)'], df3['petal width (cm)'], color='black', label='Petal width 3')

plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='purple', marker='+', label='centroid')


plt.xlabel('Petal Length {cm}')
plt.ylabel('Petal width {cm}')
plt.legend()
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['petal length (cm)', 'petal length (cm)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of square error')
plt.plot(k_rng, sse)
```
