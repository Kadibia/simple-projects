```python
import pandas as pd
df = pd.read_csv('heart.csv')
df.head()
```

```python
df['target'] = df['HeartDisease']
df = df.drop('HeartDisease', axis='columns')
df
```

```python
df.ChestPainType.describe()
```

```python
df = pd.get_dummies(df, drop_first=True )
df
```

```python
df.dtypes
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['Age', 'RestingBP', 'Cholesterol', 'FastingBS', 'MaxHR', 'Oldpeak', 'Sex_M', 'ChestPainType_ATA', 'ChestPainType_NAP', 'ChestPainType_TA', 'RestingECG_Normal', 'RestingECG_ST', 'ExerciseAngina_Y', 'ST_Slope_Flat', 'ST_Slope_Up']], df.target, test_size=0.2, random_state=42)
```

```python
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
print(X_train_scaled)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
lr = LogisticRegression(max_iter=1000)
svm = SVC()
rfc = RandomForestClassifier()
```

```python
lr.fit(X_train_scaled, y_train)
```

```python
svm.fit(X_train_scaled, y_train)
```

```python
rfc.fit(X_train_scaled, y_train)
```

```python
lr.score(X_train_scaled, y_train)
```

```python
lr.score(X_test_scaled, y_test)
```

```python
lr.predict(X_test_scaled)
```

```python
svm.score(X_train_scaled, y_train)
```

```python
svm.predict(X_test_scaled)
```

```python
rfc.score(X_train_scaled, y_train)
```
